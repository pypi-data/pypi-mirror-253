```python
from wdm_wavelet.wdm import WDM
```

```python
from pycbc.waveform import get_td_waveform

hp, hc = get_td_waveform(approximant="IMRPhenomTPHM",
                         mass1=20,
                         mass2=20,
                         spin1z=0.9,
                         spin2z=0.4,
                         inclination=1.23,
                         coa_phase=2.45,
                         distance=100,
                         delta_t=1.0/2048,
                         f_lower=20)

hp.plot()
```

```python
import matplotlib.pyplot as plt

wdm = WDM(32, 64, 6, 10)
m_L, nWWS, pWDM = wdm.t2w(hp.data)
plt.imshow(pWDM[0].T, aspect='auto', origin='lower')
plt.colorbar()
```
